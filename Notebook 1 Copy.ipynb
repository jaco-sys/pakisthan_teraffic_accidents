```python
from pyspark.sql.functions import *

df_metrics=df_rename.withColumn("Toatl_patients",col('Killed')+ col('Injured'))

display(df_metrics)
```

```python
df_renamed.columns
```

```python
from pyspark.sql.functions import when, round

# Create 'Fatality_Rate' and round to 2 decimal places
# We use 'when' to avoid errors if Total_Accidents is 0
df_metrics = df_metrics.withColumn("Fatality_Rate", 
    when(col("Total_Accidents") > 0, round((col("Killed") / col("Total_Accidents")) * 100, 2))
    .otherwise(0))
```

```python
# Categorize the severity of the year/region
df_metrics = df_metrics.withColumn("Severity_Level", 
    when(col("Killed") > 500, "High Risk")
    .when(col("Killed") > 100, "Medium Risk")
    .otherwise("Low Risk"))
```

```python
from pyspark.sql.functions import col, round, when

# 1. Injury to Death Ratio
df_final = df_metrics.withColumn("Injury_Death_Ratio", 
    when(col("Killed") > 0, round(col("Injured") / col("Killed"), 2))
    .otherwise(col("Injured")))

# 2. Weighted Severity Score (Death = 5 points, Injury = 1 point)
df_final = df_final.withColumn("Severity_Score", 
    (col("Killed") * 5) + (col("Injured") * 1))

display(df_final)
```

```python
# Replaces nulls with 0 in every numeric column automatically
df_clean = df_final.fillna(0)
```

```python
df_clean.write.format("delta").mode("overwrite").saveAsTable("gold_road_safety_final")
```
